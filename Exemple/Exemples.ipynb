# Axe1: les indicateurs techniques

La librairie contient en total 14 indicateurs techniques :

                     - Moyenne mobile simple	  
                     - Moyenne mobile exponentielle 	   
                     - Moving Average Convergence Divergence 
                     - Relative Strength Index
                     - Bandes de Bollinger 	 
                     - Momentum	
                     - Rate of Change 
                     - Stochastique 	 
                     - On-Balance
                     - Williams	%R
                     - Money Flow Index 
                     - Chaikin Oscillator
                     - Negative Volume Index
                     - Positive Volume index  
Dans cette partie on  va vous expliquer comment utiliser la librairie pour calculer les indicateurs mentionnés ci-dessus 


**1- Moyenne Mobile Simple :**

       - Syntaxe: mms(df,w)
       - Les paramètres: - df : le vecteur des prix 
                         - w  : l'ordre de la moyenne mobile 
       - Exemple : si on veut calculer la moyenne mobile des prix de clôture sur une période de 20 jours, le syntaxe à
                   utiliser est mms(close, 20) 

```python
# Exemple de la moyenne mobile simple 
moy=at.mms(close,52)
moy.tail(5)
```

2- Moyenne Mobile Exponentielle :

       - Syntaxe: mme(df,w)
       - Les paramètres: - df : le vecteur des prix 
                         - w  : l'ordre de la moyenne mobile 
       - Exemple : si on veut calculer la moyenne mobile exponentielle des prix de clôture sur une période de 20 jours, le                    syntaxe à utiliser est mme(close, 20)

```python
#Exemple de la moyenne mobile simple 
moy=at.mme(close,50)
moy.tail(5)
```

3-Moving Average Convergence Divegence-MACD (plus le signal line : la moyenne mobile de la MACD) 

       - Syntaxe: macd(df, ws, wl, wsig)
       - Les paramètres: - df : le vecteur des prix
                         - ws : l'ordre de court terme
                         - wl : l'ordre de long terme 
                         - wsig : l'ordre de la signal line
       - Exemple : La MACD des prix de clôture si l'ordre de court terme est 12, long terme 26 et l'ordre de signal line 
                   est 9 donc le syntaxe est  macd(close,12,26,9)

```python
#MACD
mcd=at.macd(close,12,26,9)
mcd.tail(5)
```

4- Relative Strength Index- RSI

             - Syntaxe : rsi(df,n)
             - Paramètre : - df : le vecteur des prix de clôture 
                           - n  : l'ordre 
             - Exemple : Pour pouvoir calculer le RSI d'ordre de 14 la syntaxe à utiliser est rsi(close,14) 
      

```python
#RSI
rsi=at.rsi(close,14)
rsi.tail(5)
```

5-Bandes de Bollinger 

         - Syntaxe : bollinger(df,w,k) 
         - Paramètre:  - df : les prix de clôture
                       - w  : Ordre de la moyenne mobile
                       - k  : le multiplicateur (pour construire les  enveloppes supérieure et inférieure) 
         - Exemple :les bandes de bollinger dont les enveloppes sont deux fois l'écart-type sont obtenus en utilisant
                    le syntaxe suivant bollinger(close,20,2) 
                    
         * Pour l'output, on adopte la convention suivante :
                BBDOWN : l'enveloppe inférieure
                BBUP   : l'enveloppe supérieure 
                BBMID  : La moyenne mobile 

```python
#Bande de Bollinger
bb=at.bollinger(close,20,2)
bb.tail(5)
```

6-Momentum 

            - Syntaxe : momentum(df,w,wsig=9)
            - Paramètre :  - df :les prix de clôture
                           - w  : l'ordre du momentume
                           - wsig : l'ordre de la moyenne mobile sur le momentum le signal line 
            - Exemple : le momentum pour l'ordre 12 et pour le signal d'ordre 9, le syntaxe à utiliser
                        est momentum(close,12,9)

```python
# Momentum
mm=at.momentum(close,12,9)
mm.tail(5)
```

7- Rate of Change - ROC

            - Syntaxe : rate_of_change(df,w)
            - Paramètres:  - df : Les prix de clôture
                           - w  : L'ordre 
            - Exemple : ROC des prix sur une période w= 12 est calculée par rate_of_change(close,12)

```python
# Rate of Change
roc=at.rate_of_change(close,12)
roc.tail(5)
```

**8- Stochastique: %K et %D**

             - Syntaxe : stochastique(df ,n , w)
             - Paramètres:  - df : Les prix de clôture
                            - n  : l'ordre de calcule de %K
                            - w  : l'ordre de calcule de %D
             - Exemple: les stochastique %K sur 14 et %D sur 3 est donnés par stochastique(close ,14, 3)

```python
# Stochastique
stoc=at.stochastique(close,14,3)
stoc.tail(5)
```

**9- On Balance Volume: OBV** 

            - Syntaxe:  obv(df,vol)
            - Paramètre:  - df : Les prix de clôture
                          - vol: Les volume 

```python
#OBV
obv=at.obv(close,volume)
obv.tail(5)
```

**11- Williams %R**
   
         - Syntaxe : williams(df,n)
         - Paramètres:  - df : Les prix de clôture
                        - n  : L'ordre 
         - Exemple   williams(close,14) williams de l'ordre de 14 

```python
# Williams %R
wl=at.williams(close,14)
wl.tail(5)
```

**12-Money Flow Index**

         - Syntaxe: MFI(close,volume,high,low,n)
         - Paramètre:  - close : Les prix de clôture
                       - volume : les volumes
                       - high  : Les prix les plus haut durant chaque séance
                       - low : Les prix les plus bas durant chaque séance 

```python
# Money Flow Index
mfi=at.MFI(close,volume,high,low,14)
mfi.tail(5)
```

**13-Chaikin Oscillator**

           - Syntaxe:  cho(close,volume,high,low,n,ws,wl)
           - Paramètres:   - close : Les prix de clôture
                           - volume : les volumes
                           - high  : Les prix les plus haut durant chaque séance
                           - low : Les prix les plus bas durant chaque séance
                           - n : l'ordre de l'ADL (voir la formule de calcule de Chaikin Oscillator)
                           - ws : l'ordre de court terme
                           - wl : l'ordre de long terme 

```python
# Chaikin oscillator
cho=at.cho(close,volume,high,low,9,3,10)
cho.tail(5)
```

**13- Negative Volume Index- NVI**

            -Syntaxe : nvi(close,volume)
            -Paramètre:  - close : Les prix de clôture
                         - volume : Les volumes 

```python
# NVI
nv=at.nvi(close,volume)
nv.tail(5)
```

**13- Positive Volume Index- PVI**

            -Syntaxe : pvi(close,volume)
            -Paramètre:  - close : Les prix de clôture
                         - volume : Les volumes 

```python
# PVI
pv=at.nvi(close,volume)
pv.tail(5)
```

# Axe 2: les graphiques indicateurs techniques

Donc les fonction existantes dans cette libairie sont les suivantes: 
   - Moyenne mobile simple	     : plot_mms (df, n) 
   - Moyenne mobile exponentielle: plot_mme(df, n)

   - RSI	                     : plot_rsi(df, n)
   - Bande de Bollinger          : plot_bollinger( df, w, k )
   - Momentum	                 : plot_momentum(df, w, wsig )
   - ROC 	                     : plot_roc(df, w) 
   - Stochastique 	             : plot_stochastique(df, n, w,) 
   - OBV	                     : plot_obv(df, vol)
   - Williams	                 : plot_williams(df, n)
   - MFI	                     : plot_MFI(close, volume, high, low, n)
   - CHO	                     : plot_cho(close, volume, high, low, n, ws, wl )
   - NVI	                     : plot_nvi(close, volume)
   - PVI	                     : plot_pvi(close, volume)
